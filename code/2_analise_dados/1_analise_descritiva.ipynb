Resumo Estatístico

```python
df_resumo_estatistico = pd.DataFrame(columns=['Segmento', 'Média', 'Valor Mínino', 'Valor Máximo', 'Desvio Padrão'])

for segmento in df_media_segm.columns:
    
    if df_resumo_estatistico.index.isna().all():

        df_resumo_estatistico = pd.DataFrame({
            'Segmento' : segmento,
            'Média' : df_media_segm[segmento].describe()['mean'], 
            'Valor Mínino' : df_media_segm[segmento].describe()['min'], 
            '1o Quartil (%25)' : df_media_segm[segmento].describe()['25%'],
            'Mediana' : df_media_segm[segmento].describe()['50%'],
            '3o Quartil (75%)' : df_media_segm[segmento].describe()['75%'],
            'Valor Máximo' : df_media_segm[segmento].describe()['max'],
            'Desvio Padrão' : df_media_segm[segmento].describe()['std']},
            index=[segmento])
    
    else:
        
        df_res_estat_segm  = pd.DataFrame({
            'Segmento' : segmento,
            'Média' : df_media_segm[segmento].describe()['mean'], 
            'Valor Mínino' : df_media_segm[segmento].describe()['min'], 
            '1o Quartil (%25)' : df_media_segm[segmento].describe()['25%'],
            'Mediana' : df_media_segm[segmento].describe()['50%'],
            '3o Quartil (75%)' : df_media_segm[segmento].describe()['75%'],
            'Valor Máximo' : df_media_segm[segmento].describe()['max'],
            'Desvio Padrão' : df_media_segm[segmento].describe()['std']},
            index=[segmento])
    
        df_resumo_estatistico = pd.concat([df_resumo_estatistico, df_res_estat_segm], ignore_index=True)

df_resumo_estatistico.set_index("Segmento", inplace=True)
df_resumo_estatistico.index = df_resumo_estatistico.index.astype(str)

df_resumo_estatistico.head()
```

```python
idx_max_med = df_resumo_estatistico["Média"].idxmax()
val_max_med = df_resumo_estatistico["Média"].max()
print(f"O maior valor médio foi {val_max_med}, referente ao segmente {idx_max_med}.")
idx_min_med = df_resumo_estatistico["Média"].idxmin()
val_min_med = df_resumo_estatistico["Média"].min()
print(f"O menor valor médio foi {val_min_med}, referente ao segmente {idx_min_med}.")
```

```python
# O menor desvio padrão indica a menor volatilidade e, consequentemente o menor risco? (Melhor maior diferença positiva entre valor final e inicial.) 
idx_max_std = df_resumo_estatistico["Desvio Padrão"].idxmax()
val_max_std = df_resumo_estatistico["Desvio Padrão"].max()
print(f"O maior desvio padrão médio foi {val_max_std}, referente ao segmento {idx_max_std}.")
idx_min_std = df_resumo_estatistico["Desvio Padrão"].idxmin()
val_min_std = df_resumo_estatistico["Desvio Padrão"].min()
print(f"O menor desvio padrão médio foi {val_min_std}, referente ao segmento {idx_min_std}.")
```

Análise das séries temporais

Lembrete: segmentos com maior tendência de alta possuem o maior potencial de retorno.

```python
import numpy as np
from statsmodels.tsa.stattools import acf

df_tend = pd.DataFrame(index=df_media_segm.index)

for segmento in df_media_segm.columns:

    # 22 dias de cotação por mês / 252 dias de cotação por ano / df_media_segm.shape[0]//2 maior período possível para decomposição da base de dados
    
    df_tend[f"{segmento}"] = seasonal_decompose(df_media_segm[f"{segmento}"],
                                                period = 252,
                                                model = "additive").trend
df_tend.dropna(inplace=True)
# df_tend
df_tend["Acessórios"].plot()
```

Lembrete: a sazonalidade de cada segmento indica os possíveis bons momentos de compra e venda de ativos do segmento.

```python
df_saz = pd.DataFrame(index=df_media_segm.index)

for segmento in df_media_segm.columns:
    # 22 dias de cotação por mês
    # 252 dias de cotação por ano
    # maior período possível para decomposição da base de dados
    df_saz[f"{segmento}"] = seasonal_decompose(df_media_segm[f"{segmento}"],
                                period = df_media_segm.shape[0]//2,
                                model = "additive").seasonal
                            # df_media_segm.shape[0]//2
df_saz.dropna(inplace=True)
# df_saz
df_saz["Acessórios"].plot()
```

Análise de correlação entre as sazonalidades dos segmentos.

```python
mtx_corr_saz = df_saz.corr()
mtx_corr_saz.head()
```

Segmente com maior correlação negativa entre suas sazonalidades pode indicar boas alternativas para 

Proteção de carteira

Rebalanceamento de carteira

Estratégias de swing trade

```python
list_segm_ref = []
list_segm_corr_neg = []
list_val_corr_neg = []
    
for segmento in mtx_corr_saz.columns:

    list_segm_ref.append(segmento)
    list_segm_corr_neg.append(mtx_corr_saz[segmento].idxmin())
    list_val_corr_neg.append(mtx_corr_saz[segmento].sort_values(ascending=True).iloc[0])

df_corr_neg = pd.DataFrame({
    'Segmento 1' : list_segm_ref,
    'Segmento 2' : list_segm_corr_neg,
    'Correlação' : list_val_corr_neg
})

df_corr_neg.set_index('Segmento 1', inplace=True)
df_corr_neg = df_corr_neg.sort_values('Correlação', ascending=True)
display(df_corr_neg)


    
    # print(f"segmento analisado = {segm_ref}\nsegmento com maior correlação negativa = {segm_corr_neg}\nCorrelação = {val_corr_neg}\n")
```

Lembrete: segmentos cujo resíduo apresenta menor desvio padrão pode indicar os setores que apresentam menor risco. 

```python
df_resid = pd.DataFrame(index=df_media_segm.index)

for segmento in df_media_segm.columns:
    
    # 22 dias de cotação por mês / 252 dias de cotação por ano / df_media_segm.shape[0]//2 maior período possível para decomposição da base de dados

    df_resid[f"{segmento}"] = seasonal_decompose(df_media_segm[f"{segmento}"],
                                period = 252,
                                model = "additive").resid
                            
df_resid.dropna(inplace=True)
# df_resid
df_resid["Acessórios"].plot()
```
